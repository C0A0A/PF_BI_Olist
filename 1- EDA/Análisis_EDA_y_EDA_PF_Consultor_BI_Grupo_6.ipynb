A tener en cuenta

**Nulos**

Si la base está en alcance del proyecto, se debe proceder a tratar los datos nulos del conjunto. Se debe validar:
-	Si existe una columna en la misma base que refiera a otra base donde se pudiera obtener el dato faltante e imputarlo si fuera posible., se realiza la imputación.
-	Si no existe la columna, se valida:
o	si la cantidad de registros con valores nulos no es representativa del conjunto, se procederá a eliminarse los registros del conjunto.
o	Si la cantidad de registros con valores nulos es representativa del conjunto, pero la columna no aporta información al objetivo del trabajo, entonces se elimina la columna.
Si la base no está en alcance del proyecto, se desestima la base.

**Normalización de las tablas**

Algunas tablas no tienen una buena normalización realizada, dado que existe información extra a la necesaria. Por ejemplo, customer tiene su id, id_zip_code, ciudad y estado. Si estuviera bien normalizada la tabla de ciudades y estados, se podría dejar únicamente la referencia al id_zip_code y todo el resto de la información de las ciudades visualizarse en la tabla de ciudades y estados.

**Limpieza de datos**

Verificar que los datos numéricos tengan el mismo formato para poder realizar operaciones entre si y utilizarlos posteriormente en un modelo. Modificar formato de ser necesario.
Verificar que los datos categóricos tengan datos apropiados para la información de la columna. Se detectó un correo electrónico en la variable ciudad, en la tabla de Sellers.
